```python
import pandas as pd
import numpy as np
import os

data_filepath = '../../Data/ds-capstone-dataset'

data_files = [file for file in os.listdir(data_filepath) if file[-4:] != '.csv']

labels = pd.read_csv(f"{data_filepath}/label_1980_2018_rawyield.csv")

# Rename first column
labels = labels.rename(columns={"Unnamed: 0":"year_index"})

print(data_files)
```

```python
labels
```

```python
all_avg_data = labels.copy()

for file in data_files:
    feature = file[2:-19]
    data = np.load(f'{data_filepath}/{file}', allow_pickle=True)
    data_df = pd.DataFrame(data)
    all_avg_data[feature] = data_df.mean(axis=1)

all_avg_data.to_csv(f"./Data/all_feature_data_avg.csv")
```
