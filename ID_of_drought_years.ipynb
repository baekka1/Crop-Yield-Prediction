```python
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/Data Science Capstone/label_1980_2018_rawyield.csv'
df = pd.read_csv(file_path)
```

```python
mean_yield_by_year = df.groupby('year')['yield'].mean()
```

```python
plt.subplots(figsize=(10, 6))
plt.plot(mean_yield_by_year.index, mean_yield_by_year.values, marker='o')
plt.xlabel("Year")
plt.ylabel("Average yield of all counties")
plt.title("Average Corn Yield Over Time")
plt.grid(True)
plt.xticks(mean_yield_by_year.index, rotation=45, fontsize=8)
for x, y in zip(mean_yield_by_year.index, mean_yield_by_year.values):
    plt.text(x, y, f'{y:.2f}', ha='left', va='bottom', fontsize=6, rotation=45)
plt.show()
```
