```python
import pandas as pd
import numpy as np

file_name = 'all_feature_data_avg.csv'
data = pd.read_csv(file_name)
display(data.head())
```

```python
data['year'] = pd.to_datetime(data['year'], format='%Y')
data = data.set_index('year_index')
data = data.sort_index()
display(data.head(10))
display(data.isnull().sum())
```

```python
from statsmodels.tsa.ar_model import AutoReg
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

target_variable = 'yield'
time_series_data = data[target_variable]

# Split data into training and testing sets(80% train, 20% test)
train_size = int(len(time_series_data) * 0.8)
train_data, test_data = time_series_data[0:train_size], time_series_data[train_size:]

# Using lag of 1 as a starting point
ar_model = AutoReg(train_data, lags=1).fit()

start_index = len(train_data)
end_index = len(time_series_data) - 1
ar_predictions = ar_model.predict(start=start_index, end=end_index)

ar_results = pd.DataFrame({'Actual': test_data, 'AR_Predictions': ar_predictions.values}, index=test_data.index)
display(ar_results.head())
rmse = np.sqrt(mean_squared_error(test_data.values, ar_predictions.values))
print(f'RMSE for AR model: {rmse}')
```

## ARIMA Model

```python
from statsmodels.tsa.arima.model import ARIMA

# p: number of lag observations (AR order) - starting with 5
# d: the number of times the raw observations are differenced (Integrated order) - starting with 1 for differencing
# q: the size of the moving average window (MA order) - starting with 0
arima_order = (5, 1, 0)
arima_model = ARIMA(train_data, order=arima_order)
arima_results = arima_model.fit()

arima_forecast = arima_results.forecast(steps=len(test_data))

arima_forecast.index = test_data.index
arima_results_df = pd.DataFrame({'Actual': test_data, 'ARIMA_Predictions': arima_forecast})
display(arima_results_df.head())
rmse_arima = np.sqrt(mean_squared_error(test_data, arima_forecast))
print(f'RMSE for ARIMA model: {rmse_arima}')
```

## Differenced VAR model



```python
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

var_variables = ['yield', 'tmmx', 'rmax', 'vs', 'sph', 'srad', 'vpd', 'rmin', 'pr', 'tmmn', 'th']
data_cleaned = data[var_variables].copy() # Create a copy to avoid SettingWithCopyWarning

differenced_data = pd.DataFrame(index=data_cleaned.index)
differenced_data['yield'] = data_cleaned['yield'].diff()

for col in var_variables:
    if col != 'yield':
        differenced_data[col] = data_cleaned[col]

differenced_data = differenced_data.dropna()


differenced_data_reset = differenced_data.reset_index(drop=True)


train_size_var = int(len(differenced_data_reset) * 0.8)
train_data_var, test_data_var = differenced_data_reset[0:train_size_var], differenced_data_reset[train_size_var:]

# Using lag of 1 as a starting point
var_model = VAR(train_data_var)
var_results = var_model.fit(1)

var_forecast_steps = len(test_data_var)
var_forecast = var_results.forecast(train_data_var.values[-var_results.k_ar:], steps=var_forecast_steps)

var_forecast_df = pd.DataFrame(var_forecast, columns=differenced_data.columns)

last_train_value_yield = data_cleaned['yield'].iloc[train_size-1]
actual_var_eval = data_cleaned['yield'].iloc[train_size:]

var_forecast_yield_diff = var_forecast_df['yield']
var_forecast_yield = var_forecast_yield_diff.cumsum() + last_train_value_yield

original_test_index = differenced_data.iloc[train_size_var:].index[:len(var_forecast_yield)].copy()


actual_var_eval_aligned = actual_var_eval.iloc[:len(var_forecast_yield)]

var_results_yield = pd.DataFrame({'Actual': actual_var_eval_aligned.values, 'VAR_Predictions': var_forecast_yield.values}, index=original_test_index)
display(var_results_yield.head())
rmse_var = np.sqrt(mean_squared_error(actual_var_eval_aligned.values, var_forecast_yield.values))
print(f'RMSE for Differenced VAR model (yield): {rmse_var}')
```

## SARIMA Model


```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd


train_data_reset_index = train_data.reset_index(drop=True)


sarima_order = (1, 1, 1)
seasonal_order = (1, 1, 1, 210) # growing season of 210 days

sarima_model = SARIMAX(train_data_reset_index,
                       order=sarima_order,
                       seasonal_order=seasonal_order,
                       enforce_stationarity=False,
                       enforce_invertibility=False)

sarima_results = sarima_model.fit()

sarima_forecast = sarima_results.forecast(steps=len(test_data))

rmse_sarima = np.sqrt(mean_squared_error(test_data.values, sarima_forecast.values))
print(f'RMSE for SARIMA model: {rmse_sarima}')

sarima_forecast.index = test_data.index
sarima_results_df = pd.DataFrame({'Actual': test_data, 'SARIMA_Predictions': sarima_forecast})
display(sarima_results_df.head())
```
